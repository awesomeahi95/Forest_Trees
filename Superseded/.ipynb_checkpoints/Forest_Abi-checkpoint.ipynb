There are 7 different categories of trees on this forest. We are only interested on type 2 = "Lodgepole Pine"

```python
# We transform the categories to convert it to a binary combination of type 2= yes or the other types = 0 
df_forest['Pine'] = df_forest.Cover_Type.apply(lambda x: 1 if x == 2 else 0)
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(df_forest['Pine'])
ax.set_xticklabels(["Other","Lodgepole Pine"])
```

Both categories are balanced
